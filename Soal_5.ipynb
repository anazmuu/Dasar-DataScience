```python
karyawan_a = 20
karyawan_b = 14
karyawan_c = 28
karyawan_d = 42 
karyawan_e = 18

gaji_a = 300000
gaji_b = 360000
gaji_c = 475000
gaji_d = 580000
gaji_e = 400000
asuransi = 75000

print("Soal 1")
pengeluaran_a = (gaji_a * karyawan_a) + (karyawan_a * asuransi)
pengeluaran_b = (gaji_b * karyawan_b) + (karyawan_b * asuransi)
pengeluaran_c = (gaji_c * karyawan_c) + (karyawan_c * asuransi)
pengeluaran_d = (gaji_d * karyawan_d) + (karyawan_d * asuransi)
pengeluaran_e = (gaji_e * karyawan_e) + (karyawan_e * asuransi)
print("total pengeluaran gaji dan asuransi perusahaan A : rp. " + str(pengeluaran_a))
print("total pengeluaran gaji dan asuransi perusahaan B : rp. " + str(pengeluaran_b))
print("total pengeluaran gaji dan asuransi perusahaan C : rp. " + str(pengeluaran_c))
print("total pengeluaran gaji dan asuransi perusahaan D : rp. " + str(pengeluaran_d))
print("total pengeluaran gaji dan asuransi perusahaan E : rp. " + str(pengeluaran_e))

print("Soal 2")
pajak = 0.03
aset_a = 280000000
aset_b = 310000000
aset_c = 420000000
aset_d = 550000000
aset_e = 340000000
laba_a = 32500000
laba_b = 31000000
laba_c = 42000000
laba_d = 51500000
laba_e = 27000000
utang_a = 0.08
utang_b = 0.04
utang_c = 0.12
utang_d = 0.11
utang_e = 0.03

total_pu_a = (pajak * laba_a) + (utang_a * aset_a)
total_pu_b = (pajak * laba_b) + (utang_b * aset_b)
total_pu_c = (pajak * laba_c) + (utang_c * aset_c)
total_pu_d = (pajak * laba_d) + (utang_d * aset_d)
total_pu_e = (pajak * laba_e) + (utang_e * aset_e)
print("Maka total pajak dan utang a adalah : rp. " + str(total_pu_a))
print("Maka total pajak dan utang b adalah : rp. " + str(total_pu_b))
print("Maka total pajak dan utang c adalah : rp. " + str(total_pu_c))
print("Maka total pajak dan utang d adalah : rp. " + str(total_pu_d))
print("Maka total pajak dan utang e adalah : rp. " + str(total_pu_e))

import matplotlib.pyplot as plt
#pie chart
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
plt.title("Data Pengeluaran pajak dan gaji")
label_makanan = ['A-Xis', 'B-Log', 'C-in', 'D-namic', 'E-psilon']
jml = [total_pu_a, total_pu_b, total_pu_c, total_pu_d, total_pu_e ]
ax.pie(jml,labels=label_makanan,autopct='%1.2f%%')
plt.show()

#cara menampilkan grafik BAR
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
jurusan = ['A-Xis', 'B-Log', 'C-in', 'D-namic', 'E-psilon']
siswa = [total_pu_a, total_pu_b, total_pu_c, total_pu_d, total_pu_e ]
c = ['red', 'green', 'blue', 'yellow', 'orange']
plt.title("Data Pengeluaran pajak dan gaji")
plt.bar(jurusan, siswa, color = c)
plt.show()

print("Soal 3")
csr_kurang = 0.025
csr_lebih = 0.035

csr_a = ((csr_kurang) * aset_a)
csr_b = ((csr_lebih) * aset_b)
csr_c = ((csr_lebih) * aset_c)
csr_d = ((csr_lebih) * aset_d)
csr_e = ((csr_lebih) * aset_e)
print("Maka CSR perusahaan a adalah : rp. " + str(csr_a))
print("Maka CSR perusahaan b adalah : rp. " + str(csr_b))
print("Maka CSR perusahaan c adalah : rp. " + str(csr_c))
print("Maka CSR perusahaan d adalah : rp. " + str(csr_d))
print("Maka CSR perusahaan e adalah : rp. " + str(csr_e))
#pie chart
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
plt.title("Data Pengeluaran pajak dan gaji")
label_makanan = ['A-Xis', 'B-Log', 'C-in', 'D-namic', 'E-psilon']
jml = [csr_a, csr_b, csr_c, csr_d, csr_e]
ax.pie(jml,labels=label_makanan,autopct='%1.2f%%')
plt.show()

#cara menampilkan grafik BAR
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
jurusan = ['A-Xis', 'B-Log', 'C-in', 'D-namic', 'E-psilon']
siswa = [csr_a, csr_b, csr_c, csr_d, csr_e]
c = ['red', 'green', 'blue', 'yellow', 'orange']
plt.title("Data Pengeluaran pajak dan gaji")
plt.bar(jurusan, siswa, color = c)
plt.show()

print("Soal 4")
total_laba = (12 * laba_a) + (12 * laba_b) + (12 * laba_c) + (12 * laba_d) + (12 * laba_e)
print("total laba semua perusahaan selama 1 tahun adalah : Rp. " + str(total_laba))

print("Soal 5")
pendapatan_bersih_a = (12 * laba_a) - csr_a - (12 * pengeluaran_a) - total_pu_a
pendapatan_bersih_b = (12 * laba_b) - csr_b - (12 * pengeluaran_b) - total_pu_b
pendapatan_bersih_c = (12 * laba_c) - csr_c - (12 * pengeluaran_c) - total_pu_c
pendapatan_bersih_d = (12 * laba_d) - csr_d - (12 * pengeluaran_d) - total_pu_d
pendapatan_bersih_e = (12 * laba_e) - csr_e - (12 * pengeluaran_e) - total_pu_e
print("total pendapatan bersih pertahun perusahaan a : rp. " + str(pendapatan_bersih_a))
print("total pendapatan bersih pertahun perusahaan b : rp. " + str(pendapatan_bersih_b))
print("total pendapatan bersih pertahun perusahaan c : rp. " + str(pendapatan_bersih_c))
print("total pendapatan bersih pertahun perusahaan d : rp. " + str(pendapatan_bersih_d))
print("total pendapatan bersih pertahun perusahaan e : rp. " + str(pendapatan_bersih_e))





```
